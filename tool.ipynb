```python
dt = get_data('pycaret_downloads')
```

```python
dt.info()
```

```python
import pandas as pd
```

```python
dt['Date']= pd.to_datetime(dt['Date'])
```

```python
dt.info()
```

```python
dt.set_index('Date',drop=True, inplace=True)
```

```python
dt
```

```python
dt.plot();
```

```python
from pycaret.time_series import *
```

```python
ts=setup(dt,fh=15,fold_strategy='sliding')
```

```python
best_model=compare_models(n_select=3)
```

```python
arima=create_model('arima')
```

```python
plot_model(arima)
```

```python
arima_tuned=tune_model(arima)
```

```python
plot_model(arima_tuned)
```

```python
prophet=create_model('prophet')
```

```python
plot_model(prophet)
```

```python
prophet_tuned=tune_model(prophet)
```

```python
plot_model(prophet_tuned)
```

```python
blender=blend_models(best_model)
```

```python
plot_model(blender)
```

```python
final_model=finalize_model(blender)
```

```python
final_model
```

```python
plot_model(final_model,plot='forecast')
```

```python
final_model_arima=finalize_model(arima)
```

```python
final_model_arima
```

```python
plot_model(final_model_arima,plot='forecast')
```

```python
predict_model(final_model,fh=15)
```

```python
predict_model(final_model_arima)
```
